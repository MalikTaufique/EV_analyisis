```python
# Example handling missing values

```

```python
Data.fillna({
    'kwhTotal':Data['kwhTotal'].median(),
    'platform':Data['platform'].mode()[0],
}, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x=Data["kwhTotal"])
plt.title("Outlier Detection for kWh Consumed")
plt.show()
```

```python
# Remove outliers using the IQR method
Q1 = Data["kwhTotal"].quantile(0.25)
Q3 = Data["kwhTotal"].quantile(0.75)
IQR = Q3 - Q1

# Define acceptable range
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter data
Data = Data[(Data["kwhTotal"] >= lower_bound) & (Data["kwhTotal"] <= upper_bound)]
```

```python
Data =pd.get_dummies(Data,columns=["platform","weekday"],drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
cols_to_scale =["kwhTotal", "chargeTimeHrs", "distance"]

# apply MinMax scaling
Data[cols_to_scale] = scaler.fit_transform(Data[cols_to_scale])
```

```python
# Distribution of Energy Consumption (kWh)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.histplot(Data["kwhTotal"], bins=30, kde=True, color="blue")
plt.title("Distribution of Energy Consumption (kWh)")
plt.xlabel("kWh Consumed")
plt.ylabel("Frequency")
plt.show()
```

```python
#Charging Sessions by Hour of the Day
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=Data["chargeTimeHrs"], y=Data["kwhTotal"], alpha=0.6)
plt.title("Charging Duration vs Energy Consumption")
plt.xlabel("Charging Duration (Hours)")
plt.ylabel("kWh Consumed")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=Data["locationId"], order=Data["locationId"].value_counts().index[:10], palette="magma")
plt.title("Top 10 Charging Locations by Demand")
plt.xlabel("Location ID")
plt.ylabel("Number of Charging Sessions")
plt.xticks(rotation=45)
plt.show()
```

```python
# Select only numeric columns for correlation
numeric_df = Data.select_dtypes(include=["number"])

# Generate heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(y=Data["kwhTotal"], color="red")
plt.title("Box Plot of Energy Consumption (kWh)")
plt.ylabel("kWh Consumed")
plt.show()
```
